```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

df = pd.read_csv('superloja.csv')

display(df.head())

df.info()

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Sales', bins=50, kde=True, color='skyblue')
plt.title('Distribuio de Vendas na loja')
plt.xlabel('Valor das Vendas ($)')
plt.ylabel('Contagem')
plt.xlim(0, 1000) 
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Category', y='Sales', palette='Set2')
plt.title('Distribuio de Vendas por Categoria de Produto')
plt.yscale('log') 
plt.show()

plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='Profit', hue='Region', fill=True, common_norm=False, palette='crest')
plt.title('Densidade do Lucro por Regio')
plt.xlim(-200, 200)
plt.show()

plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Discount', hue='Segment', multiple='stack', bins=10, palette='pastel')
plt.title('Distribuio de Descontos: Consumidor vs Corporativo vs Home Office')
plt.xlabel('Desconto Aplicado')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Region', y="Sales", hue='Category', palette='Set3')
plt.title('Vendas por Regio e Categoria')
plt.xlabel('Regio')
plt.yscale('log')
plt.legend(title="Categoria")
plt.show()

plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Ship Mode', y='Profit', palette='muted')
plt.title('Distribuio do Lucro por Modo de Envio')
plt.ylim(-100, 200)
plt.show()
```
