```python
#Para rodar a aplicação

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")

df = pd.read_csv('titanic.csv')

#Mostra as primeiras 5 linhas formadas 
display(df.head())
#Dica: No VS code,use "display()" ao inves de "print()" para tabela bonita

#Verificando tipos de dados e nulos
df.info()

#Análise da Idade
plt.figure(figsize=(10,6))
sns.histplot(data = df,x='Age',bins=30,kde=True,color='skyblue')
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.xlabel('Contagem')
plt.show()

#Comparação de Tarifa por Classe
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Pclass', y='Fare', palette='Set2')
plt.title('Preço da Passagem por classe(Bilhetes)')
plt.show()

#Apenas a linha de densidade(sem barras)
plt.figure(figsize=(10,6))
sns.kdeplot(data = df,x='Age',hue='2urvived',fill=True,common_norm=False,palette='crest')
plt.title('Densidade de Idade: Quem sobreviveu (1) vs Morreu (0)')
plt.show()

#Histograma de idade separado por Sexo
plt.figure(figsize=(10,6))
sns.histplot(data = df,x='Age',hue='Sex',multiple='stack',bins=25,palette='pastel')
plt.title('Distribuição de Idade: Homens vs Mulheres')
plt.xlabel('Idade')
plt.show()

#Preço X Classe, separando sobrevivencia
plt.figure(figsize=(10,6))
sns.boxplot(data = df,x='Pclass',y="Fare",hue='2urvived',palette='Set3')
plt.title('Impacto do Preço na Sobrevivencia por Classe')
plt.xlabel('Preço Passageiro')
plt.legend(title="Sobreviveu?")
plt.show()

#Violin Plot: A evolução do box plot
plt.figure(figsize=(10,6))
sns.violinplot(data = df,x='Pclass',y="Age",split=True,hue='Sex',palette='muted')
plt.title('Distribuição de Idade por Classe e Sexo')
plt.show()
```
